【問題1】100日目の米粒の数

```python
grain_0 = 1

list_n_grains = [] #ある日に貰う米の数のリスト (リストの長さはn_daysに等しい)
list_total_grains = [] #ある日までに累計で貰う米の数のリスト (リストの長さはn_daysに等しい)

for i in range(100):
    n_grains = grain_0 * (2**i)
    list_n_grains.append(n_grains)
    list_n_total_grains = sum(list_n_grains)
    list_total_grains.append(list_n_total_grains)

print("米粒数 : {}粒".format(n_grains))
len(list_n_grains)
```

```python
len(list_total_grains)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# ある日に貰う米の数の推移
plt.title("Number of grains in a day")
plt.xlabel("number of the day")
plt.ylabel("Number of the grain [粒]")
plt.tick_params(labelsize=15)
plt.plot(list_n_grains, color='red')
plt.show()
```

```python
#ある日までに累計で貰う米の数のリスト
plt.title("Number of grains in a day")
plt.xlabel("number of the day")
plt.ylabel("Number of the grain [粒]")
plt.tick_params(labelsize=15)
plt.plot(list_total_grains, color='red')
plt.show()
```

【問題2】100日目以外の米粒の数

```python
#曽呂利新左衛門の米の逸話における、日にちごとの米の粒の数を計算する関数。
#n_days: int
#    米を貰う日数 (default : 100)

def compute_sorori_shinzaemon(n_days):
  list_n_grains = []
  list_total_grains = []
  total_grains = 0  # 総米粒の数を初期化

  for i in range(n_days):
        n_grains = 2 ** i  # n日目の米粒の数
        list_n_grains.append(n_grains)

        total_grains += n_grains  # 総数を累積
        list_total_grains.append(total_grains)

  return list_n_grains, list_total_grains
```

```python
# 関数を実行
list_n_grains, list_total_grains = compute_sorori_shinzaemon(10)
print(f"{n_days}日目の米粒の数: {list_n_grains[-1]}")
print(f"{n_days}日目までの累積の米粒の総数: {list_total_grains[-1]}")
```

```python
n_days = 30
list_n_grains, list_total_grains = compute_sorori_shinzaemon(30)
print(f"{n_days}日目の米粒の数: {list_n_grains[-1]}")
print(f"{n_days}日目までの累積の米粒の総数: {list_total_grains[-1]}")
```

【問題3】貰える米で何人が何日生活できるか

ChatGPTが提示する情報から厚生労働省のサイトに飛び調べると、日本人1人が1日に消費するお米の量はおよそ150g。
また、ジャポニカ米は1粒0.02gとのこと。
ここから、7,500粒/日/人として計算する。

```python
def days_stock(n_grains, population):
  return n_grains / (7500 * population)
```

```python
days_stock(1073741823, 800)
```

100日間続けた場合、全世界の人口80億人で食しても膨大な年数がかかることになってしまったため、【問題2】最後の30日間でもらえる米粒の総数、107,341,823粒を800人の村の人々で分配して食べる計算を行った。
