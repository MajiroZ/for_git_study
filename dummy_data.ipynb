## 【問題1】乱数の作成

```python
import numpy as np

mean = (-3, 0)
# 共分散行列
cov = [[1.0, 0.8], [0.8, 1.0]]
# 乱数の数
size = 500

# 多変量正規分布から乱数を生成
samples = np.random.multivariate_normal(mean, cov, size)
np.random.seed(0)
print(samples[:5])
```

##【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 散布図を表示する
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tick_params(labelsize=15)  # 軸の値に関する設定
# Access the individual columns of the samples array for x and y coordinates
plt.scatter(samples[:, 0], samples[:, 1])  # samples[:, 0] provides the x-coordinates and samples[:, 1] provides the y-coordinates
plt.show()
```

##【問題3】ヒストグラムによる可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline

# ヒストグラムを表示する
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.tick_params(labelsize=15)

# `bins` 引数
plt.hist(samples[:, 0], bins=50)

plt.xlim(-6, 3)  # 必要なら調整
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# ヒストグラムを表示する
plt.title("Scatter")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.tick_params(labelsize=15)

# `bins` 引数
plt.hist(samples[:, 1], bins=50)

plt.xlim(-6, 3)  # 必要なら調整
plt.show()
```

##【問題4】データの追加

```python
mean = (0, -3)
# 共分散行列
cov = [[1.0, 0.8], [0.8, 1.0]]
# 乱数の数
size = 500

# 多変量正規分布から乱数を生成
samples1 = np.random.multivariate_normal(mean, cov, size)
np.random.seed(0)
print(samples1[:5])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 散布図を表示する
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.tick_params(labelsize=15)  # 軸の値に関する設定

# samples の散布図
plt.scatter(samples[:, 0], samples[:, 1], label="0", color="blue", alpha=0.7)

# samples1 の散布図
plt.scatter(samples1[:, 0], samples1[:, 1], label="1", color="red", alpha=0.7)

# 凡例を追加
plt.legend()

plt.show()
```

##【問題5】データの結合

```python
np.concatenate([samples, samples1])
```

##【問題6】ラベル付け

```python
combined_samples = np.concatenate([samples, samples1])
labels = np.array([0] * len(samples) + [1] * len(samples1))
combined_data = np.column_stack((combined_samples, labels))
print(combined_data)
```
