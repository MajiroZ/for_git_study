##**【問題1】コンペティション内容の確認**

###Home Credit Default Riskでは、何を学習し何を予測するのか
Home Creditでは、信用履歴がない人々に金融サービスを提供し、安全で前向きな借入体験を通じて金融包摂を推進している。
通信データや取引情報などの代替データを活用して返済能力を予測する。 顧客が返済可能な条件でローンを提供することで、未利用層を支援する。

###どのようなファイルを作りKaggleに提出するのか
ファイルは、テストセットの各 SK_ID_CURR についてTARGET変数の確率を予測したものを記載し、ヘッダを含める必要がある。

（例）
```
SK_ID_CURR, TARGET
10001, 0.1
10005, 0.9
100013, 0.2
```



###提出されたものはどういった指標値で評価されるのか
提出されたデータは、予測された確率と観測されたターゲットとの間のROC曲線下面積で評価される。

````python
# データの概観を把握する
import pandas as pd
df_original = pd.read_csv('/application_train.csv')

df_original.head()
````

````python
df_original.describe()
````

````python
# 5つ以上欠損値がある特徴量（列）を削除する

# 各列の欠損値の個数を数える
missing_counts = df_original.isnull().sum()

# 5つ以上の欠損値のある列を特定する
columns_to_drop = missing_counts[missing_counts >= 5].index

# 削除
df = df_original.drop(columns=columns_to_drop)

# 欠損値のある行を特定する
lines_to_drop = df[df.isnull().any(axis=1)].index

# 削除
df = df.drop(index=lines_to_drop)

print(df.head())
````

````python
# 各クライアントのローン返済確率（0以上1以下）を目的変数として追加し
# application_testのデータでも予測・検証してみる

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score


# 目的変数と説明変数に分ける
y = df["TARGET"]
X = df.drop("TARGET", axis=1)

# 数値データのみを使用する場合
X = X.select_dtypes(include=['number'])

# 学習データと検証データに分割
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)

# ロジスティック回帰モデルのインスタンスを作成
model = LogisticRegression(random_state=0, max_iter=1000) # max_iterを増やす

# モデルを学習データにフィットさせる
model.fit(X_train, y_train)

# 検証データで予測
y_pred_val = model.predict_proba(X_val)[:, 1] # 確率値を取得

# ROC AUCを計算
roc_auc = roc_auc_score(y_val, y_pred_val)
print(f"ROC AUC (Validation): {roc_auc}")

# テストデータの読み込み
df_test = pd.read_csv('/application_test.csv')

# テストデータの前処理（学習データと同じ処理を行う）
df_test = df_test.drop(columns=columns_to_drop)

# テストデータの欠損値処理
for col in df_test.columns:
    if df_test[col].isnull().any():
        if df_test[col].dtype == 'object':
            df_test[col] = df_test[col].fillna(df_test[col].mode()[0])
        else:
            df_test[col] = df_test[col].fillna(df_test[col].mean())

# 数値データのみを使用
X_test = df_test.select_dtypes(include=['number'])

# 同一の列のみを使用するように調整
X_test = X_test[X_train.columns]

# テストデータで予測
y_pred_test = model.predict_proba(X_test)[:, 1]

# 提出用データフレームを作成
submission = pd.DataFrame({'SK_ID_CURR': df_test['SK_ID_CURR'], 'TARGET': y_pred_test})

# CSVファイルとして保存
submission.to_csv('submission.csv', index=False)
print(submission.head())
print("Submission file saved successfully!")
````

````python
# ロジスティック回帰モデルにbureau.csvのデータを追加して学習・検証

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier

# データの読み込み
df_original = pd.read_csv('/application_train.csv')
bureau = pd.read_csv('/bureau.csv')

# bureauデータの前処理
# 平均のローン金額を計算して新しい特徴量を追加)
bureau_agg = bureau.groupby('SK_ID_CURR')['AMT_CREDIT_SUM'].mean().reset_index()
bureau_agg.rename(columns={'AMT_CREDIT_SUM': 'BUREAU_AMT_CREDIT_SUM_MEAN'}, inplace=True)
df_original = df_original.merge(bureau_agg, on='SK_ID_CURR', how='left')

# 欠損値処理
for col in df_original.columns:
    if df_original[col].isnull().any():
        if df_original[col].dtype == 'object':
            df_original[col] = df_original[col].fillna(df_original[col].mode()[0])
        else:
            df_original[col] = df_original[col].fillna(df_original[col].mean())

# 5つ以上欠損値がある特徴量（列）を削除する
missing_counts = df_original.isnull().sum()
columns_to_drop = missing_counts[missing_counts >= 5].index
df = df_original.drop(columns=columns_to_drop)

# 欠損値のある行を削除する
df = df.dropna()

# 目的変数と説明変数に分ける
y = df["TARGET"]
X = df.drop("TARGET", axis=1)

# 数値データのみを使用
X = X.select_dtypes(include=['number'])

# 学習データと検証データに分割
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)

# ロジスティック回帰モデル
model = LogisticRegression(random_state=0, max_iter=1000)
model.fit(X_train, y_train)
y_pred_val = model.predict_proba(X_val)[:, 1]
roc_auc = roc_auc_score(y_val, y_pred_val)
print(f"Logistic Regression ROC AUC (Validation): {roc_auc}")


# ランダムフォレストモデル
model_rf = RandomForestClassifier(random_state=0, n_estimators=100)
model_rf.fit(X_train, y_train)
y_pred_val_rf = model_rf.predict_proba(X_val)[:, 1]
roc_auc_rf = roc_auc_score(y_val, y_pred_val_rf)
print(f"Random Forest ROC AUC (Validation): {roc_auc_rf}")
````

````python
# application_testに関して推定を行い提出データファイルを作成・保存・表示する

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier

# データの読み込み (パスは適宜変更してください)
df_original = pd.read_csv('/application_train.csv')
bureau = pd.read_csv('/bureau.csv')
previous_application = pd.read_csv('/previous_application.csv')
pos_cash_balance = pd.read_csv('/POS_CASH_balance.csv')
credit_card_balance = pd.read_csv('/credit_card_balance.csv')


# 特徴量エンジニアリング (既存のコードを元に、必要な部分を抜粋・修正)
bureau_agg = bureau.groupby('SK_ID_CURR')['AMT_CREDIT_SUM'].mean().reset_index()
bureau_agg.rename(columns={'AMT_CREDIT_SUM': 'BUREAU_AMT_CREDIT_SUM_MEAN'}, inplace=True)
df_original = df_original.merge(bureau_agg, on='SK_ID_CURR', how='left')

previous_agg = previous_application.groupby('SK_ID_CURR')['AMT_CREDIT'].mean().reset_index()
previous_agg.rename(columns={'AMT_CREDIT': 'PREVIOUS_AMT_CREDIT_MEAN'}, inplace=True)
df_original = df_original.merge(previous_agg, on='SK_ID_CURR', how='left')

pos_cash_agg = pos_cash_balance.groupby('SK_ID_CURR')['CNT_INSTALMENT'].mean().reset_index()
pos_cash_agg.rename(columns={'CNT_INSTALMENT': 'POS_CASH_CNT_INSTALMENT_MEAN'}, inplace=True)
df_original = df_original.merge(pos_cash_agg, on='SK_ID_CURR', how='left')

credit_card_agg = credit_card_balance.groupby('SK_ID_CURR')['AMT_BALANCE'].mean().reset_index()
credit_card_agg.rename(columns={'AMT_BALANCE': 'CREDIT_CARD_AMT_BALANCE_MEAN'}, inplace=True)
df_original = df_original.merge(credit_card_agg, on='SK_ID_CURR', how='left')


# 欠損値処理と特徴量選択 (数値データのみ使用)
for col in df_original.columns:
    if df_original[col].isnull().any():
        if df_original[col].dtype == 'object':
            df_original[col] = df_original[col].fillna(df_original[col].mode()[0])
        else:
            df_original[col] = df_original[col].fillna(df_original[col].mean())


# 5つ以上欠損値がある特徴量（列）を削除する
missing_counts = df_original.isnull().sum()
columns_to_drop = missing_counts[missing_counts >= 5].index
df = df_original.drop(columns=columns_to_drop)

# 欠損値のある行を削除する
df = df.dropna()

# 目的変数と説明変数に分ける
y = df["TARGET"]
X = df.drop("TARGET", axis=1)

# カテゴリカル変数の処理 (factorizeを使用)
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = pd.factorize(X[col])[0]


# 学習データと検証データに分割
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)

# ランダムフォレストモデルを使用 (より精度の高いモデルを選択)
model_rf = RandomForestClassifier(random_state=0, n_estimators=100)
model_rf.fit(X_train, y_train)
y_pred_val_rf = model_rf.predict_proba(X_val)[:, 1]
roc_auc_rf = roc_auc_score(y_val, y_pred_val_rf)
print(f"Random Forest ROC AUC (Validation): {roc_auc_rf}")


# テストデータの読み込みと前処理
df_test = pd.read_csv('/application_test.csv')

# テストデータにも同じ処理を行う
df_test = df_test.merge(bureau_agg, on='SK_ID_CURR', how='left')
df_test = df_test.merge(previous_agg, on='SK_ID_CURR', how='left')
df_test = df_test.merge(pos_cash_agg, on='SK_ID_CURR', how='left')
df_test = df_test.merge(credit_card_agg, on='SK_ID_CURR', how='left')


for col in df_test.columns:
    if df_test[col].isnull().any():
        if df_test[col].dtype == 'object':
            df_test[col] = df_test[col].fillna(df_test[col].mode()[0])
        else:
            df_test[col] = df_test[col].fillna(df_test[col].mean())

df_test = df_test.drop(columns=columns_to_drop, errors='ignore')

# カテゴリカル変数の処理
for col in df_test.columns:
  if df_test[col].dtype == 'object':
    df_test[col] = pd.factorize(df_test[col])[0]

X_test = df_test[X_train.columns] # 学習データと同じカラムのみを使用

# テストデータで予測
y_pred_test = model_rf.predict_proba(X_test)[:, 1]

# 提出用データフレームを作成
submission = pd.DataFrame({'SK_ID_CURR': df_test['SK_ID_CURR'], 'TARGET': y_pred_test})

# CSVファイルとして保存
submission.to_csv('submission_rf_final.csv', index=False)
print(submission.head())
print("Submission file saved successfully!")
````
