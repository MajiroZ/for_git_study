## 【問題】栗まんじゅうが太陽系を覆う日

栗きんとん すやのホームページによると、この和菓子屋で製作されている栗まんじゅうは、
大きさ(1コ)：60mm x 40mm x 30mm
とのこと。この値 72㎤ を解答に使用する。

また、太陽系の断面の形状は楕円形か、球形に広がっているのか、実はまだ明らかになっていないという。
以前は冥王星の軌道を太陽系外縁と定義していたこともあるが、今は太陽風が到達して周囲の星間物質と混じり合うヘリオポーズ（100-200au）までの範囲を太陽圏として太陽系と定義することもあるという。
今回はこの定義（公益社団法人 日本天文学会『天文学時点』より）を採用し、
100au = 100x149,597,871km
これを半径、高さを太陽の直径 1,392,700km の円盤として、9.7917*(10^26)㎦ を太陽系体積として用いる。

```python
# 太陽系の体積
import math
((100*149597871)*1000)**2 * math.pi * 1392700*1000
```

```python
list_volume = []
def baibain(VOLUME, target):
  for i in range(10000):
    VOLUME = VOLUME * 2
    list_volume.append(VOLUME)
    if VOLUME > target :
      break
  return "体積 : {}[㎥]".format(VOLUME), "回数 : {}回".format(i+1), "所要時間 : {}[分]".format((i+1)*5)

VOLUME = 72*(10**(-6))  # 栗まんじゅう1個の体積
target = 9.791703948168613e+35

baibain(VOLUME, target)
```

したがって、134回繰り返すのにかかる時間、670分が、太陽系が栗まんじゅうで覆われるまでにかかる時間といえる。

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Volume of Kurimanju over Time")
plt.xlabel("Elapsed Time [min]")  # x軸のラベル「経過時間[分]」
plt.ylabel("Volume [m**3]")
plt.tick_params(labelsize=15)  # 軸の値に関する設定
plt.plot([i * 5 for i in range(len(list_volume))], list_volume, color='red')  # x軸：経過時間[分]
plt.show()
```
